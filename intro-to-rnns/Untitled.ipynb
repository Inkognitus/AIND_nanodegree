```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(w, s1, i):
    if i == 0:
        return s1
    s = f(w, s1, i-1)    
    return w * s - w * s ** 2
```

```python
w = 1; s1 = 0.5
x = np.arange(0,50)
y = [f(w, s1, i) for i in x]
```

```python
plt.plot(y)
```
